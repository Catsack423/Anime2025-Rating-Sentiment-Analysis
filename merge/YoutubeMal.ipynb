```python
import json
import pandas as pd

# โหลดไฟล์
with open("../commentMyanimelist/comment.json", "r", encoding="utf-8") as f:
    comment_data = json.load(f)

with open("../commentYoutube/anime_comments_cleaned_promax.json", "r", encoding="utf-8") as f:
    promax_data = json.load(f)

with open("../commentAnilist/cleaned_anime_comments_pro.json", "r", encoding="utf-8") as f:
    pro_data = json.load(f)


# แปลงเป็น DataFrame
df_comment = pd.DataFrame(comment_data)
df_promax = pd.DataFrame(promax_data)
df_pro = pd.DataFrame(pro_data)


# ปรับชื่อ column ให้เหมือนกัน
df_comment = df_comment.rename(columns={
    "title": "anime_name",
    "text_c": "comment"
})

df_promax = df_promax.rename(columns={
    "author": "user_c",
    "comment_text": "comment"
})

df_pro = df_pro.rename(columns={
    "anime":"anime_name",
    "user": "user_c",
    "text": "comment"
})


# เลือก column ที่ต้องใช้
df_comment = df_comment[["anime_name", "user_c", "comment"]]
df_promax = df_promax[["anime_name", "user_c", "comment"]]
df_pro = df_pro[["anime_name", "user_c", "comment"]]


# รวมทั้งหมด
df_all = pd.concat([df_comment, df_promax, df_pro], ignore_index=True)


# สร้าง ID
df_all.insert(0, "ID", range(1, len(df_all) + 1))

# บันทึกไฟล์
#df_all.to_csv("YoutubeMal_comments.csv", index=False, encoding="utf-8")

# แสดงตัวอย่าง
print(df_comment.info())
print(df_promax.info())
print(df_pro.info())

display(df_all)

```
