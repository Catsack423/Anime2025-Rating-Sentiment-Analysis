```python
import json
import pandas as pd


with open("../commentYoutube/anime_comments_cleaned_promax.json", "r", encoding="utf-8") as f:
    youtube_data = json.load(f)

with open("../animelist/anime_2025_by_season.json", "r", encoding="utf-8") as f:
    anime_data = json.load(f)



df_youtube = pd.DataFrame(youtube_data)

df_youtube = df_youtube.rename(columns={
    "author": "user_c",
    "comment_text": "comment"
})

df_youtube["anime_name"] = df_youtube["anime_name"].str.lower().str.strip()

print("YouTube comments:", len(df_youtube))


anime_map = {}

for season, animes in anime_data["seasons"].items():
    for anime in animes:

        mal_id = anime["mal_id"]

        titles = []

        if anime["title"]:
            titles.append(anime["title"])

        if anime["title_english"]:
            titles.append(anime["title_english"])

        titles.extend(anime["title_synonyms"])

        for t in anime["titles"]:
            titles.append(t["title"])

        for title in titles:
            key = title.lower().strip()
            if key not in anime_map:
                anime_map[key] = mal_id


print("Total mapped titles:", len(anime_map))



df_youtube["malID"] = df_youtube["anime_name"].map(anime_map)



df_final = df_youtube[["malID", "anime_name", "user_c", "comment"]]

df_final.insert(0, "ID", range(1, len(df_final)+1))



df_final.to_json(
    "Youtube_comments_cleaned_final.json",
    orient="records",
    force_ascii=False,
    indent=2
)

print("\nFinal Dataset Info")
print(df_final.info())

print("\nMissing malID:", df_final["malID"].isna().sum())

display(df_final.head())
```
